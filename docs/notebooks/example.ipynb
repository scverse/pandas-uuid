# Examples

```python
from uuid import uuid4

import pandas as pd

from pandas_uuid import UuidDtype, UuidExtensionArray
```

```python
UuidExtensionArray(values=[0, uuid4()])
```

```python
s = pd.Series([uuid4()], dtype=UuidDtype(), name="s")
s
```

```python
pd.DataFrame(s)
```

## pyarrow integration

convert from {mod}`pandas` to {mod}`pyarrow` automatically â€¦

```python
import pyarrow as pa

arr = pa.array(s.array)
arr
```

and back manually (there is no registry, so neither {class}`pyarrow.UuidArray` nor {mod}`pandas` knows, only our {class}`~pandas_uuid.UuidDtype`):

```python
pd.Series(arr, dtype=UuidDtype())
```
